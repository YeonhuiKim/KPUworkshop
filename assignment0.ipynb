```python
x=np.arange(0, 3*np.pi, 0.1)
y=np.sin(x)
plt.xlim(0,3*np.pi)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sinusoidal curve(s)')

x2=np.arange(0,3*np.pi,0.1)
y2=np.cos(x2)

plt.plot(x,y)
plt.plot(x2,y2)
plt.show()
```

```python
x=np.arange(0,3*np.pi, 0.1)
y_sin=np.sin(x)

fig, axes=plt.subplots(nrows=2)
axes[0].plot(x,y_sin)
axes[0].set_xlim(0,3*np.pi)
axes[0].set_title('Sine')

y_cos=np.cos(x)

axes[1].plot(x,y_cos)
axes[1].set_xlim(0,3*np.pi)
axes[1].set_title('Cosine')

plt.tight_layout()
fig.show()
```
